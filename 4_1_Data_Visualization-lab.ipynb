# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
mpl.style.use('ggplot')
```

```python
# your code goes here
QUERY = """SELECT * FROM master"""

df_master = pd.read_sql_query(QUERY,conn)

plt.hist(df_master['ConvertedComp'], bins = 20)
plt.title('Histogram of annual compensation in $')
plt.xlabel('Compensation ($)')
plt.ylabel('Frequency')
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
avg_age = df_master['Age'].mean() #30.778894788947888
df_master_age_NoNan = df_master
df_master_age_NoNan['Age'].fillna(avg_age, inplace=True)
df_master_age_NoNan['Age'] = df_master_age_NoNan['Age'].fillna(avg_age)
df_master_age_NoNan['Age'].isnull().sum()
plt.boxplot(df_master_age_NoNan['Age'])
plt.title('Age summary')
plt.ylabel('Age')
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
df_master.plot(kind='scatter', x='WorkWeekHrs', y='Age', alpha=0.35)
plt.title('Scatter plot of relation between age and week hours')
plt.ylabel('Age (y)')
plt.xlabel('Work week (h)')
```

Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
QUERY = """
SELECT * FROM DatabaseDesireNextYear
"""
df_dbdesire = pd.read_sql_query(QUERY, conn)
df_dbdesire.head()
```

```python
QUERY = """
SELECT DatabaseDesireNextYear,COUNT(*) as Count
FROM DatabaseDesireNextYear
group by DatabaseDesireNextYear
order by Count desc
"""

df_dbdesirecount = pd.read_sql_query(QUERY, conn)
df_dbdesirecount
```

```python
df_dbdesirecount_topfive = df_dbdesirecount[0:5]
df_dbdesirecount_topfive
df_dbdesirecount_topfive = df_dbdesirecount_topfive.set_index('DatabaseDesireNextYear')
df_dbdesirecount_topfive
```

```python
explode_list = [0.1,0,0,0,0]
df_dbdesirecount_topfive['Count'].plot(kind='pie',
                                       figsize=(13, 6),
                                       autopct='%1.1f%%',
                                       startangle=90,
                                       pctdistance=0.8,    # the ratio between the center of each pie slice and the start of the text generated by autopct 
                                       labels=df_dbdesirecount_topfive.index,
                                       labeldistance=1.1,
                                       explode=explode_list
                                      )

# scale the title up by 12% to match pctdistance
plt.title('Top 5 Databases desired to be used for the following year', y=1.12) 

plt.axis('equal') 

# add legend
plt.legend(labels=df_dbdesirecount_topfive.index, loc='upper left') 

plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age FROM master
WHERE Age BETWEEN 30 AND 35
GROUP BY Age
"""
df_master_hours = pd.read_sql_query(QUERY,conn)

df_master_hours.plot(x='Age', kind='bar', stacked=True,
                     title='Work week and Code Revision hours by age between 30-35',
                    figsize=(10,6))
plt.ylabel('Hours')
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
#QUERY = """
#SELECT ConvertedComp, Age,
#        PERCENTILE_CONT(0.5)
#            WITHIN GROUP (ORDER BY ConvertedComp)
#            OVER (PARTITION BY Age)
#            AS Median_ConvertedComp
#FROM master
#WHERE Age BETWEEN 45 AND 60
#ORDER BY Age, ConvertedComp
#"""

QUERY = """
SELECT ConvertedComp, Age FROM master
WHERE Age BETWEEN 45 AND 60
ORDER BY Age
"""
df_master_ConvComp = pd.read_sql_query(QUERY, conn)
df_master_ConvComp.head()
```

```python
df_master_ConvComp_median = df_master_ConvComp.groupby('Age')['ConvertedComp'].median().to_frame()
df_master_ConvComp_median
```

```python
df_master_ConvComp_median.plot(kind='line', figsize=(10,6))
plt.title('Median values of annual converted compensation ($) per age')
plt.ylabel('Converted Compensation ($)')
```
